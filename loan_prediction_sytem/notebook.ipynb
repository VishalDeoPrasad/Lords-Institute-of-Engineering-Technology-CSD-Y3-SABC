```python
df.drop(['Loan_ID'], axis=1, inplace=True)
```

```python
cat_cols = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area']
cat_cols
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
df.Gender = le1.fit_transform(df.Gender)
```

```python
le2 = LabelEncoder()
df.Married = le2.fit_transform(df.Married)
```

```python
le3 = LabelEncoder()
df.Education = le3.fit_transform(df.Education)
```

```python
le4 = LabelEncoder()
df.Self_Employed = le4.fit_transform(df.Self_Employed)
```

```python
le5 = LabelEncoder()
df.Property_Area = le5.fit_transform(df.Property_Area)
```

```python
le6 = LabelEncoder()
df.Loan_Status = le6.fit_transform(df.Loan_Status)
```

```python
le7 = LabelEncoder()
df.Dependents = le7.fit_transform(df.Dependents)
```

```python
encoders = {
    'Gender' : le1,
    'Married' : le2,
    'Dependents': le7,
    'Education' : le3,
    'Self_Employed' : le4,
    'Property_Area' : le5,
}
```

```python
import joblib
joblib.dump(encoders, "encoders.joblib")
```

```python
df.head()
```

```python
df.Dependents.unique()
```

```python
x = df.drop(['Loan_Status'], axis=1)
y = df.Loan_Status
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(max_iter=5000)
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier()
model3.fit(x_train, y_train)
y_pred = model3.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model4 = KNeighborsClassifier()
model4.fit(x_train, y_train)
y_pred = model4.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
model5 = SVC()
model5.fit(x_train, y_train)
y_pred = model5.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model6 = GradientBoostingClassifier()
model6.fit(x_train, y_train)
y_pred = model6.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
joblib.dump(model6, "loan_model.joblib")
```
